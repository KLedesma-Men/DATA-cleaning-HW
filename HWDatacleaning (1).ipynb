```python
#%% Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#%% Importing Data
flights_data = pd.read_csv('/Users/karlyn/Downloads/data/flights.csv')  
flights_data.head(10)
weather_data_pd = pd.read_csv('/Users/karlyn/Downloads/data/weather.csv')
weather_data_np = weather_data_pd.to_numpy()
#%% Pandas Data Filtering/Sorting Question Answering
#use flights_data

#Question 1 How many flights were there from JFK to SLC? Int
q_1 = len(flights_data[(flights_data['origin'] == 'JFK') & (flights_data['dest'] == 'SLC')])
print("The number of flights from JFK to SLC were:",q_1)

#Question 2 How many airlines fly to SLC? Should be int
q_2 = flights_data[flights_data['dest'] == 'SLC']['carrier'].nunique()
print("The number of airlines flown into slc were:",q_2)

#Question 3 What is the average arrival delay for flights to RDU? float
#filter info first 
RDU_flights = flights_data[flights_data['dest'] == 'RDU']
#now calculation 
q_3 = RDU_flights['arr_delay'].mean()
print("The average delay is:",q_3)

#Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float
#filter flights first 
SEA_flights = flights_data[flights_data['dest'] == 'SEA']
#now filter flights from LGA and JFK 
NYC_flights  = SEA_flights[SEA_flights['origin'].isin(['LGA', 'JFK'])] 
#count total flights 
total_flights = SEA_flights.shape[0]
#count NYC flights 
NYC_count_flights = NYC_flights.shape[0]
#now the proportion 
proportion_of_flights = NYC_count_flights / total_flights if total_flights > 0 else 0.0
print("The proportion of flights are:",proportion_of_flights)

#Question 5 Which date has the largest average depature delay? Pd slice with date and float
#please make date a column. Preferred format is 2013/1/1 (y/m/d)
#print(flights_data.columns)
#creating new data column 
flights_data['date'] = pd.to_datetime(flights_data[['year', 'month', 'day']])
#to the desired format 
flights_data['date'] = flights_data['date'].dt.strftime('%Y/%m/%d')
#calculating average delay 
avg_delay = flights_data.groupby('date')['dep_delay'].mean()
#finding the date with the largest average departure date
max_delay = avg_delay.idxmax()
print("The date with the largest average departure delay is:" ,max_delay)

#Question 6 Which date has the largest average arrival delay? pd slice with date and float
avg_arrival_delay = flights_data.groupby('date')['arr_delay'].mean()
#finding the date
arrival_delay_date = avg_arrival_delay.idxmax()
print("The date with the largest average arrival delay is:" ,arrival_delay_date)

#Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed
#speed = distance/airtime
#filter the flights from LGA to JFK 
filtered_flights = flights_data[(flights_data['origin'].isin(['LGA', 'JFK'])) & (flights_data['year'] == 2013)]
#calculate speed with formula
filtered_flights['speed'] = filtered_flights['distance'] / filtered_flights['air_time']
#finding the max speed 
fastest_flight = filtered_flights.loc[filtered_flights['speed'].idxmax(), ['tailnum', 'speed']]
print(f"The fastest flight was {fastest_flight['tailnum']} with a speed of {fastest_flight['speed']:.2f} mph")

#Question 8 Replace all nans in the weather pd dataframe with 0s. Pd with no nans
#replace nans with 0s
weather_data_pd.fillna(0, inplace=True)
print(weather_data_pd)

#%% Numpy Data Filtering/Sorting Question Answering
#Use weather_data_np

#Question 9 How many observations were made in Feburary? Int
february_observations = flights_data[flights_data['month'] == 2].shape[0]
print(f"The number of observations in February is:{february_observations}")

#Question 10 What was the mean for humidity in February? Float
#filter weather data 
february_weather_data = weather_data_pd[weather_data_pd['month'] == 2]
#calcualte mean 
mean_humidity = february_weather_data['humid'].mean()
print(f"The mean humidity in February is: {mean_humidity}")

#Question 11 What was the std for humidity in February? Float
std_humid = february_weather_data['humid'].std()
print(f"The standard deviation of humidity in February is: {std_humid}")
```
